```python
song_artist_map = {}
def createMaps(row):
    row.artists = row.artists[1:-1].split(',')

    song_artist_map[row.id] = ({
        'song_name' : row['name'],
        'artists' : set(row.artists),
        #'artists' : set([item[0] for item in row.artists]),
        'tempo': row.tempo,
        'energy': row.energy,
        'acousticness': row.acousticness
    })

music.apply(createMaps, axis=1)
```

```python
#song_artist_map['4YmyPAgBwxfaOFq3i5nRAu']['artists'][0]
song_artist_map
```

```python
all_song_ids = np.array(list(song_artist_map.keys()))
training_ids, testing_ids = train_test_split(all_song_ids, test_size=.2)
```

```python
num_neighbors = 8
```

```python
predictions = []

for test_song_id in testing_ids:
    target_song_object = song_artist_map[test_song_id]
    test_song_sim = []

    for this_song_id in training_ids:
        this_song_object = song_artist_map[this_song_id]

        this_intersect = this_song_object['artists'].intersection(target_song_object['artists'])
        this_union = this_song_object['artists'].union(target_song_object['artists'])

        jaccard = len(this_intersect)/len(this_union)

        test_song_sim.append({
            'song': this_song_id,
            'jaccard': jaccard
        })
    sim_df = pd.DataFrame(test_song_sim, columns=['song','jaccard'])
    top_songs = sim_df.sort_values(by='jaccard', ascending=False).head(num_neighbors)
    
    relevant_tempo = []
    relevant_acous = []

    for this_song_id in top_songs['song']:
        #tempo
        relevant_tempo.append(song_artist_map[this_song_id]['tempo'])

        #acousticness
        relevant_acous.append(song_artist_map[this_song_id]['acousticness'])

    predicted_tempo = np.mean(relevant_tempo)
    predicted_acous = np.mean(relevant_acous)

    predictions.append({
        'song_id': test_song_id,
        'tempo': predicted_tempo,
        'acous': predicted_acous
    })

    #print(f"{test_song_id} is predicted at {predicted_tempo}")


    
```

```python
predict_df = pd.DataFrame(predictions).set_index('song_id')
predict_df
```

Accuracy

```python
tempo_mean_squares = []

for song_id, row in predict_df.iterrows():
    actual_tempo = song_artist_map[song_id]['tempo']
    predicted_tempo = row['tempo']

    tempo_mean_squares.append((actual_tempo - predicted_tempo) ** 2)

mse = sum(tempo_mean_squares) / len(tempo_mean_squares)


print("Tempo MSE " + str(mse))
```

```python
acous_mean_squares = []

for song_id, row in predict_df.iterrows():
    actual_acous = song_artist_map[song_id]['acousticness']
    predicted_acous = row['acous']

    acous_mean_squares.append((actual_acous - predicted_acous) ** 2)

mse = sum(acous_mean_squares) / len(acous_mean_squares)


print("Tempo MSE " + str(mse))
```
